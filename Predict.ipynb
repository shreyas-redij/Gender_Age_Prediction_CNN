```
model = keras.models.load_model('/content/drive/My Drive/Age_Prediction/VGG_Architecture.h5')
```

```
# We use a sample image and preprocess it and use the model to predict it's gender   

def loadImage(filepath):
    test_img = image.load_img(filepath, target_size=(224, 224))
    test_img = image.img_to_array(test_img)
    test_img = np.expand_dims(test_img, axis = 0)
    test_img /= 255
    return test_img

def predictGender(picture):
  prediction = model.predict(loadImage(picture))
  img = image.load_img(picture)#, target_size=(224, 224))
  plt.imshow(img)
  plt.show()

  gender = "Man" if np.argmax(prediction) == 0 else "Woman"

  print("gender: ", gender)

  # use the path of the images to make a prediction on them
picture = "/content/drive/My Drive/Age_Prediction/Test_Files/Man/man1.jpg"

predictGender(picture)
```
